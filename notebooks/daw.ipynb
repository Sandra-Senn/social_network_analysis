```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('../data/data.csv')

# Display first few rows
print(df.head())
print(df.info())
```

```python
# Remove or mark modules with "Gem. Modulbeschreibung" as prerequisite
df = df[df['Voraussetzung'] != 'Gem. Modulbeschreibung']

# Handle semester specifications with commas (e.g., "3,5")
df['Semester'] = df['Semester'].str.replace(',', ';')

# Check for missing or inconsistent Module in dependencies
missing_ids = df[df['Modul'].isna()]
print(f"Modules with missing IDs: {len(missing_ids)}")

# Identify dependency references that don't exist in Modul
if 'Abhängigkeiten' in df.columns:
    valid_ids = set(df['Modul'].dropna())
    inconsistent = df[~df['Abhängigkeiten'].isin(valid_ids) & df['Abhängigkeiten'].notna()]
    print(f"Inconsistent dependencies: {len(inconsistent)}")
    print(inconsistent[['Modul', 'Abhängigkeiten']])
```
